Отвечая на ваш вопрос: мы избавляемся после дубликатов уже после того, как все столбцы привели в единый формат, так как тогда выявится больше дубликатов. Пусть у нас в начальном датасете были две строки по  показателям "Гемоглобин (HGB)" и " (HGB ) Гемоглобин", у которых все остальные столбцы одинаковые. Когда мы переименовали весь гемоглобин в HGB - эти две строки стали дубликатами
